```python
# run langdetect on description to identify non-english description(including some special sequeces that occur)

import pandas as pd
from langdetect import detect

# Define a function to detect the language of a text
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'

# Apply the language detection function to the book description column
# This operation adds a column that shows the language - could be "en" or anything else
books_with_desc['language'] = books_with_desc['desc'].apply(detect_language)

# Filter out rows with non-English descriptions
# Store only "en" languages
books_with_desc_cleaned = books_with_desc[books_with_desc['language'] == 'en']

# Now, df_cleaned contains the DataFrame with rows containing non-English descriptions removed

books_with_desc_cleaned.shape
```

```python
# showing non-"en" languages
books_not_cleaned = books_with_desc[books_with_desc['language'] != 'en']
books_not_cleaned.head(5)
```

```python
import re

# Clean the description column from non-ASCII characters
# The description column may still contain some special characters like "ï»¿". So to remove those with the help of regex
books_with_desc_cleaned['cleaned_description'] = books_with_desc_cleaned['desc'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))
```

```python
# Same approach taken for author field as well
import re

# Clean the description column from non-ASCII characters
books_with_desc_cleaned['cleaned_author'] = books_with_desc_cleaned['ï»¿author'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))
```

```python
# Taking only the first 10000 books

books_with_desc_cleaned_trimmed = books_with_desc_cleaned.head(10000)
books_with_desc_cleaned_trimmed.shape
```

```python
# Deleting un-necessary columns

list_of_columns = ["bookformat","isbn13","link","reviews","desc","ï»¿author"]
restored_book_details = books_with_desc_cleaned_trimmed
for column_to_delete in list_of_columns:
  restored_book_details = restored_book_details.drop(column_to_delete, axis=1)

restored_book_details.head(2)
```

```python
# Renaming column names

restored_book_details.rename(columns={"cleaned_author": "author"}, inplace=True)
restored_book_details.rename(columns={"cleaned_description": "description"}, inplace=True)
restored_book_details.rename(columns={"img": "image"}, inplace=True)
restored_book_details.rename(columns={"totalratings": "total_ratings"}, inplace=True)
restored_book_details.head(2)
```

```python
# Re-ordering the columns

new_column_order = ['book_id', 'title', 'isbn', 'author', 'genre','description','image','pages','rating','total_ratings','language']
restored_book_details_rearranged = restored_book_details[new_column_order]
```

```python
restored_book_details_rearranged.head(2)
```

```python
# Function to clean up genres
# Cleaning the genre column by removing some unwanted genres
def clean_genre(genre):
    # Remove unwanted genre elements
    unwanted_elements = ["North American Hi...", "M M M", "GLBT"]
    cleaned_genres = [genre.strip() for genre in genre.split(',') if genre.strip() not in unwanted_elements]
    # Remove duplicates and return cleaned genres as a string
    return ', '.join(sorted(set(cleaned_genres)))

# Apply the clean_genre function to the genre column
restored_book_details_rearranged['genre'] = restored_book_details_rearranged['genre'].apply(clean_genre)

restored_book_details_rearranged.head(2)
```
