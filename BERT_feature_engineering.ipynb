```python
from keybert import KeyBERT
kw_model = KeyBERT()

# Takes in text input and returns a list of keywords
def get_keywords(text):

  # extract_keywords is used to extract a unigram. this function will have list of tuples.
  # tuple[0] is keyword and tuple[1] is score
    print("the text is: ",text)
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words="english")
    print("keywords are: ", keywords)

    # Take the keywords from the tuple
    keywords = " ".join([k[0] for k in keywords])
    return keywords
```

```python
data_head = cleaned_books_data.head(3)
```

```python
# Take description and apply to the keyword extraction funtion and store result in new "keywords" column
# It has to go through each description and generate the keywords so will consume like 30-60 mins for this dataset
data_head["keywords"] = data_head.description.apply(get_keywords)
```
